## Hjälpfunktioner

```python
# En fastighet (UUID) kan ägas av flera personer.
# Därför sparas varje ägares personnummer i en lista.

def get_persnr_from_uuid(uuid):
    
    # uuid_mask = fds_df['UUID-Fastighet'] == uuid
    # person_nummer = fds_df[uuid_mask]['Intressentnummer']
    # ovan i ett svep för att inte få en varning
    # se längre ner...
    
    pnr = fds_df[fds_df['UUID-Fastighet'] == uuid]['Intressentnummer']
    
    person_nummer = []
    
    for i in pnr:
        person_nummer.append(i)
        
    return person_nummer
```

```python
# Funktion som från ett givet 
# personnummer i fds-df returnerar 
# ett UUID från fds-df:

# Ett personnummer kan äga flera 
# fastigheter (UUID).

# Därför sparas varje personnummers fastigheter
# i en tuple: ([list of uuids], [len of uuids])

def get_uuid_from_persnr(persnr):
    
    member_mask = fds_df['Intressentkategori'] == 'Medlem'
    persnr_mask = fds_df['Intressentnummer'] == persnr
    
    members = fds_df[member_mask]
    
    uuid = members[persnr_mask]['UUID-Fastighet']
     
    list_of_uuids = []
    
    for i in uuid:
        list_of_uuids.append(i)
        
    return list_of_uuids, len(uuid)
```

```python
# Skicka in en lista med personnummer (string)
# för att få ut leverantörsnumret.
# En person kan ha flera fastigheter/lev.nr.:

def get_levnr_from_persnr(persnr):
    
    if type(persnr) == list:
    
        lev_nr = []
    
        for k in persnr:
            
            levnr = sir_df[sir_df['personnummer'] == str(k)]['skogsleverantörsnummer']
            lev_nr.append(levnr)
            
    else:
        
        if persnr in sir_df['personnummer'].values:
            lev_nr = sir_df[sir_df['personnummer'] == persnr]['skogsleverantörsnummer']
            
        else:
            lev_nr = []
        
    return lev_nr
```

```python
# Hämta timmer- och massavedsvolymer för
# ett leverantörsnummer:

def get_volumes(levnr):
    timber_volume = []
    pulp_volume = []
    
    for i in levnr:
        i = int(i)
        
        timber_vol = stubin_df[stubin_df['Levnr'] == i]['Timmer']
        pulp_vol = stubin_df[stubin_df['Levnr'] == i]['Massaved']
        
        timber_volume = float(timber_vol)
        pulp_volume = float(pulp_vol)
        
    return timber_volume, pulp_volume
```

```python
# Returnerar arean i hektar:

def get_sks_area(uuid):
    avv_areal = float(sks_df[sks_df['UUID'] == uuid]['AVV_AREAL'])
    return avv_areal/10000
```

```python
volume_per_hectar = 258 # https://www.skogen.se/nyheter/1630-kubikmeter-per-hektar

def calc_volume_from_area(area):
    return volume_per_hectar * area
```

```python
# 1. summera volymerna från get_volumes(levnr)

total_stubin_volumes = np.sum(volumes)
print(f'Totala volymen från Stubin är {total_stubin_volumes:,} m3fub.')
```

```python
# I verkligheten tog jag bort alla avverkningsdatum
# förutom ett i Metria-filen, vilket gör att arean 
# ovan är mycket större och kommer därför inte att
# stämma med Stubin-siffrorna.
```

```python
# En hjälpfunktion för att se om två tal är
# ungefär lika stora. "uncert" är en relativ
# tolerans, t.ex. 0.2 som det får skilja på:

def compare_volumes(estvol, truevol, uncert):
    
    return np.isclose(estvol, truevol, rtol = uncert) 
```

```python
# Jämför volymerna som uppskattats från avverkad area
# och de som är rapporterade i Stubin.
# Om de är "lika" så returneras True och omvänt.

compare_volumes(est_vol, total_stubin_volumes, 0.2)
```

```python
def loyal_member(estvol, truevol, uncert):
    if compare_volumes(estvol, truevol, uncert):
        return True
    else:
        return False
```

```python
# Är en medlem lojal som levererat 11 m3fub 
# i Stubin och som Skogsstyrelsen uppskattat
# volymen till 10 m3fub?

loyal_member(10, 11, 0.2)
```

```python
# Med givna volymer är medlemmen ovan inte att
# anses som lojal:

loyal_member(est_vol, total_stubin_volumes, 0.3)
```
